```python
df_ANN=df_for_analysis.apply(lambda x: x.fillna(x.mean()),axis=0)
df_ANN
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.model_selection import train_test_split

data=df_ANN

X = data.copy()
# En 'y' solo nos quedamos con la variable objetivo y le dropeamos la columna temperatura a X
y = X.pop('Temperature') 
print(X)
#print(y)

input_shape = [X.shape[1]]  #cantidad de columnas
print("Input shape: {}".format(input_shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(len(X_test))
print(len(X_train))
```

```python
#Usamos un multilayer perceptron con mean
#Definimos la red usada para esta task
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import BatchNormalization

# definimos el modelo de keras
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=input_shape),
    layers.Dense(8, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(1,'relu')
])
```

```python
#Definimos que tipo de "pérdida" y optimización usamos
model.compile(
    optimizer='adam',
    loss='mae',
    metrics=['mean_squared_error']
)
```

```python
#1000 epochs
import matplotlib.pyplot as plt

temp_model = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    batch_size=256,
    epochs=1000,
    verbose=0,  # turn off training log
)

temp_df = pd.DataFrame(temp_model.history)
#print(temp_df)
temp_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(temp_df['val_loss'].min()))
```

```python
import matplotlib.pyplot as plt

temp_model = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    batch_size=256,
    epochs=4100,
    verbose=0,  # turn off training log
)

temp_df = pd.DataFrame(temp_model.history)
#print(temp_df)
temp_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(temp_df['val_loss'].min()))
```

```python
import random
predictions = model.predict(X_test)               # retorna un ndarray con las predicciones

actual_values = y_test.reset_index(drop=True)     # indexes de 0..1205
actual_values

#print(f'Predicted: {predictions[429]}, Actual: {actual_values[429]}')
for x in range(25):
  index = random.randint(0, 493)
  print(f'Predicted: {predictions[index]}, Actual: {actual_values[index]}')
```
